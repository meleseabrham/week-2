```python
# Defensive load / sample fallback
"""
Cell: Load reviews dataset (auto-detect), normalize columns, 
compute rating + review stats, and generate:
 - Average rating (bar) + review volume (line)
 - Review length analysis (chars & words)
 - Rating distribution (per bank)
 - Review-length boxplot
"""

# ------------------------------------------------------------
# 1) Load dataset (auto-detect)
# ------------------------------------------------------------
try:
    df  # If already loaded above
    print("Using existing DataFrame: df")
except NameError:
    import os
    import pandas as pd

    print("Searching for review dataset...")
    candidates = [
        "data/raw_reviews.csv",
        "data/analysis/analysis_results_for_task4.csv",
        "data/cleaned_reviews.csv",
    ]

    df = None
    for p in candidates:
        if os.path.exists(p):
            df = pd.read_csv(p, parse_dates=True)
            print(f"✔ Loaded dataset from: {p}")
            break

    if df is None:
        print("⚠ No dataset found — using fallback demo DataFrame.")
        df = pd.DataFrame([
            {"bank_code": "CBE", "review_id": "r1", "rating": 5, "content": "Fast transfers, easy to use"},
            {"bank_code": "CBE", "review_id": "r2", "rating": 3, "content": "Occasional crash on login"},
            {"bank_code": "BOA", "review_id": "r3", "rating": 4, "content": "Good UI but slow sometimes"},
            {"bank_code": "BOA", "review_id": "r4", "rating": 2, "content": "OTP failed several times"},
            {"bank_code": "Dashen", "review_id": "r5", "rating": 4, "content": "Payment feature works well"},
            {"bank_code": "Dashen", "review_id": "r6", "rating": 1, "content": "App crashes when checking balance"},
        ])

# ------------------------------------------------------------
# 2) Normalize column names
# ------------------------------------------------------------
rename_map = {}

if "bank_code" not in df.columns:
    for c in ["bank", "bankName", "name"]:
        if c in df.columns:
            rename_map[c] = "bank_code"

if "review_id" not in df.columns:
    if "reviewId" in df.columns:
        rename_map["reviewId"] = "review_id"
    else:
        df = df.reset_index().rename(columns={"index": "review_id"})

if "rating" not in df.columns and "score" in df.columns:
    rename_map["score"] = "rating"

df = df.rename(columns=rename_map)

# ------------------------------------------------------------
# 3) Cleanup + ensure types
# ------------------------------------------------------------
df = df.dropna(subset=["bank_code"])  # must have bank
df["rating"] = pd.to_numeric(df.get("rating"), errors="coerce")

# Ensure content exists
if "content" not in df.columns:
    df["content"] = df.get("text", df.get("review_text", "")).fillna("").astype(str)

# ------------------------------------------------------------
# 4) Aggregate basic statistics
# ------------------------------------------------------------
if df.empty:
    print("⚠ DataFrame is empty — nothing to compute or plot.")
else:
    bank_stats = (
        df.groupby("bank_code")
        .agg(avg_rating=("rating", "mean"), review_count=("review_id", "count"))
        .reset_index()
        .sort_values("avg_rating", ascending=False)
    )

    print("✔ Computed bank-level statistics.")

# ------------------------------------------------------------
# 5) Plot: Avg rating + Review counts
# ------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

if not df.empty:
    plt.figure(figsize=(12, 6))
    ax1 = plt.gca()

    colors = sns.color_palette("tab10", n_colors=len(bank_stats))
    bars = ax1.bar(bank_stats["bank_code"], bank_stats["avg_rating"], 
                   color=colors, alpha=0.85)

    # Label bars
    for bar in bars:
        h = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width() / 2, 
                 h + 0.05, f"{h:.2f}", ha="center", fontsize=10)

    # Secondary axis for review counts
    ax2 = ax1.twinx()
    ax2.plot(bank_stats["bank_code"], bank_stats["review_count"], 
             marker="o", linewidth=1.4, color="black")

    max_count = bank_stats["review_count"].max() or 1
    for x, y in zip(bank_stats["bank_code"], bank_stats["review_count"]):
        ax2.text(x, y + max_count * 0.02, f"{y:,}", ha="center", fontsize=9)

    ax1.set_ylabel("Average Rating", fontsize=12)
    ax2.set_ylabel("Number of Reviews", fontsize=12)
    ax1.set_ylim(0, 5)

    plt.title("Bank App Performance: Avg Rating + Review Volume", fontsize=14)
    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# 6) Review length metrics
# ------------------------------------------------------------
df["review_length_chars"] = df["content"].str.len()
df["review_length_words"] = df["content"].str.split().apply(lambda x: len(x) if isinstance(x, list) else 0)

# ------------------------------------------------------------
# 7) Plot: review length, rating distribution, boxplot
# ------------------------------------------------------------
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# A) Characters distribution
sns.histplot(
    data=df, x="review_length_chars", hue="bank_code",
    bins=40, kde=True, element="step", stat="density",
    common_norm=False, ax=axes[0, 0]
)
axes[0, 0].set_title("Review Length (Characters) by Bank")

# B) Words distribution
sns.histplot(
    data=df, x="review_length_words", hue="bank_code",
    bins=40, kde=True, element="step", stat="density",
    common_norm=False, ax=axes[0, 1]
)
axes[0, 1].set_title("Review Length (Words) by Bank")

# C) Rating distribution
if "rating" in df.columns:
    sns.countplot(data=df, x="rating", hue="bank_code", ax=axes[1, 0])
    axes[1, 0].set_title("Rating Distribution by Bank")
else:
    axes[1, 0].text(0.5, 0.5, "No rating column", ha="center", va="center")

# D) Boxplot for length distribution
sns.boxplot(data=df, x="bank_code", y="review_length_chars", ax=axes[1, 1])
axes[1, 1].set_title("Review Length (Characters) — Boxplot by Bank")

plt.tight_layout()
plt.show()
```

```python
"""
Cell: parse data/analysis/analysis_report.txt and draw:
 - overall sentiment pie
 - per-bank stacked sentiment bar (percent)
 - top-themes bar chart (counts by bank)
"""
from pathlib import Path
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

report_path = Path("data/analysis/analysis_report.txt")
if report_path.exists():
    text = report_path.read_text(encoding="utf-8").splitlines()
else:
    print(f"{report_path} not found — using embedded fallback report text.")
    sample_report = """
BANK REVIEW ANALYSIS REPORT
============================================================

OVERALL SENTIMENT DISTRIBUTION:
- Positive: 52.7%
- Neutral: 38.2%
- Negative: 9.2%

============================================================
BANK: BANK OF ABYSSINIA
============================================================

SENTIMENT:
- Positive: 43.5%
- Neutral: 42.8%
- Negative: 13.8%

TOP THEMES:
- User Interface: 133 mentions
- Performance: 12 mentions

============================================================
BANK: COMMERCIAL BANK OF ETHIOPIA
============================================================

SENTIMENT:
- Positive: 55.2%
- Neutral: 39.5%
- Negative: 5.2%

TOP THEMES:
- User Interface: 124 mentions

============================================================
BANK: DASHEN BANK
============================================================

SENTIMENT:
- Positive: 59.2%
- Neutral: 32.2%
- Negative: 8.5%

TOP THEMES:
- User Interface: 169 mentions
- Performance: 33 mentions
"""
    text = sample_report.splitlines()

# Parse
overall = {}
themes_rows = []
i = 0
bank_sents = []
while i < len(text):
    line = text[i].strip()
    if line.upper().startswith("OVERALL SENTIMENT DISTRIBUTION"):
        j = i + 1
        while j < len(text) and text[j].strip() == "":
            j += 1
        for k in range(j, j + 6):
            if k >= len(text):
                break
            m = re.match(r"-\s*(Positive|Neutral|Negative):\s*([\d\.]+)%", text[k].strip(), flags=re.I)
            if m:
                overall[m.group(1).lower()] = float(m.group(2))
        i = j + 6
        continue
    m_bank = re.match(r"BANK:\s*(.+)", line, flags=re.I)
    if m_bank:
        bank_name = m_bank.group(1).strip()
        j = i + 1
        b_sent = {}
        b_themes = {}
        while j < len(text):
            l = text[j].strip()
            ms = re.match(r"-\s*(Positive|Neutral|Negative):\s*([\d\.]+)%", l, flags=re.I)
            if ms:
                b_sent[ms.group(1).lower()] = float(ms.group(2))
            mt = re.match(r"-\s*([A-Za-z0-9 &\-\_]+):\s*([\d,]+)\s*mentions", l, flags=re.I)
            if mt:
                theme = mt.group(1).strip()
                count = int(mt.group(2).replace(",", ""))
                b_themes[theme] = count
            if j > i + 1 and (l.upper().startswith("BANK:") or l.startswith("============================================================") or l == ""):
                break
            j += 1
        banksent = {
            "bank": bank_name,
            "positive": b_sent.get("positive", 0.0),
            "neutral": b_sent.get("neutral", 0.0),
            "negative": b_sent.get("negative", 0.0),
        }
        bank_sents.append(banksent)
        for th, cnt in b_themes.items():
            themes_rows.append({"bank": bank_name, "theme": th, "count": cnt})
        i = j
        continue
    i += 1

overall_series = pd.Series(overall).reindex(["positive", "neutral", "negative"]).fillna(0)
sents_df = pd.DataFrame(bank_sents).set_index("bank") if bank_sents else pd.DataFrame(columns=["positive","neutral","negative"])
themes_df = pd.DataFrame(themes_rows)

out_dir = Path("data/analysis/plots")
out_dir.mkdir(parents=True, exist_ok=True)

# Plot 1: Overall sentiment pie
plt.figure(figsize=(6,6))
colors = sns.color_palette("pastel", 3)
labels = ["Positive", "Neutral", "Negative"]
vals = overall_series.values
plt.pie(vals, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
plt.title("Overall Sentiment Distribution")
plt.tight_layout()
p1 = out_dir / "overall_sentiment_pie.png"
plt.savefig(p1, dpi=150)
plt.show()
plt.clf()

# Plot 2: Per-bank stacked bar (percent)
if not sents_df.empty:
    sents_plot = sents_df[["positive","neutral","negative"]].copy()
    sents_plot = sents_plot.loc[sorted(sents_plot.index)]
    ax = sents_plot.plot(kind="bar", stacked=True, figsize=(8,5), color=["#66c2a5","#8da0cb","#fc8d62"])
    ax.set_ylabel("Percent (%)")
    ax.set_ylim(0, 100)
    ax.set_title("Per-Bank Sentiment (percent)")
    plt.legend(title="Sentiment", bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.tight_layout()
    p2 = out_dir / "per_bank_sentiment_stacked.png"
    plt.savefig(p2, dpi=150)
    plt.show()
    plt.clf()
else:
    p2 = None
    print("No per-bank sentiment data to plot.")

# Plot 3: Top themes by bank (grouped bar)
if not themes_df.empty:
    top_themes = themes_df.groupby("theme")["count"].sum().nlargest(8).index.tolist()
    df_th = themes_df[themes_df["theme"].isin(top_themes)].pivot_table(index="theme", columns="bank", values="count", fill_value=0)
    df_th.plot(kind="bar", figsize=(10,6))
    plt.title("Top Themes (mentions) by Bank")
    plt.ylabel("Mentions (count)")
    plt.xlabel("Theme")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    p3 = out_dir / "top_themes_by_bank.png"
    plt.savefig(p3, dpi=150)
    plt.show()
else:
    p3 = None
    print("No theme data found in report to plot.")

print("Saved plots:", p1, p2, p3)
```
